```python
import csv
import glob
import os

directory = "/Users/seongyeon/Desktop/AI/AIFFEL_STUDY/DLTHON/data/01.데이터/1.Training/원천데이터/TS_01. KAKAO(1)"
output = "./data"

txt_files = os.path.join(directory, '*.txt')

normal_dict = {}
for i, txt_file in enumerate(glob.glob(txt_files)):
    f = open(txt_file, 'r')
    normal_dict[i] = f.readlines()
    f.close()
```

```python
normal_dict
```

```python
normal_df = pd.DataFrame.from_dict(data=normal_dict, orient='index')
```

```python
normal_df
```

```python
normal_df['conversation'] = normal_df.apply(lambda row: ' '.join(row.values.astype(str)), axis=1)
```
